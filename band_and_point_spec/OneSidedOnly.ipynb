```python
def mon(v,l):
#
# Calculate the general monodromy matrix with scaling l #
# and energy E
#
# Example:
# v = [1,1,0]; mon(v).expand()
# > [E^2 - E*l - 1            -E]
#   [        E - l            -1]
#
##########################
    M = identity_matrix(2)
    for vv in v:
        M = matrix(2,2,[- l*vv, -1, 1, 0])*M
    return M

def lim_pots(v):
#
# determine all unique limit potentials of a given
# periodic potential v by considering all shifts and flips
#
##########################
    v_rev = deepcopy(v)
    lim = shift_pots(v)
    v_rev.reverse()
    lim_rev = shift_pots(v_rev)
    for pot in lim_rev:
        if pot not in lim:
            lim.append(pot)
    return lim
    
def shift_pots(v):
#
# determine all unique shifts of a given periodic potential v
#
##########################
    shift = []
    for c in range(len(v)):
        if v not in shift:
            shift.append(v)
        v = right_shift(v)
    return shift
    
def right_shift(a):
    return [a[-1] , *a[:-1]]
```

```python
v = [1,0,1,1, 0,1,1,1,0]
v_rev = deepcopy(v)
v_rev.reverse()
v_list = shift_pots(v)
v_rev_list = shift_pots(v_rev)


for vv in v_list:        
    print("Potential:",vv)
    print("Monodromy matrix:")
    print(mon(vv, 1/sqrt(2)))
    print("\n")
print("------ Reversed -------")  
for vv in v_rev_list:        
    print("Potential:",vv)
    print("Monodromy matrix:")
    print(mon(vv, 1/sqrt(2)))
    print("\n")
```
